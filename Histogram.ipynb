```python
# A new function with x-axis being time and y-axis being number of particles at a given time.

# time is in microseconds: 0 to 600 microseconds
x = np.arange(0, 100, 0.04)
# y is modeled by a sin function with A = 0.25 and w = 1.53
y = 1 + (0.25*np.sin(1.53*x))

plt.title('Modified Sin Plot')
plt.xlabel('time(microseconds)')
plt.ylabel('No.of particles')
plt.plot(x, y)
plt.show()
```

```python
# Calculate the normalized integral of the above sin function with respect to time.
y_integral = [0] * len(y)
y_integral[0] = y[0]
for i in range(1, len(y)):
  integral_val = y_integral[i - 1] + y[i]
  normalized_val = integral_val / (100 - integral_val)
  y_integral[i] = integral_val

# Plot normalized integral against time
plt.title('Normalized integral plot')
plt.xlabel('time(microseconds)')
plt.ylabel('r')
plt.plot(x, y_integral)
plt.show()
```

```python
from scipy.optimize import fsolve

# returns the y-value of normalized integral
def f(x):
    rounded_x = round(x)
    return y_integral[rounded_x]

# returns corresponding x-value given the y-value
def get_corresponding_x(desired_y):
  x_solution = fsolve(lambda x: f(x) - desired_y, 0)
  return x_solution[0]
```
